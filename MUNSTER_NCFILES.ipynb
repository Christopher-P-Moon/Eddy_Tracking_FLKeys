```python
lon_min, lon_max = -83.0, -80.0
lat_min, lat_max = 24.0, 26.0

eddy_data_dir = '/content/drive/MyDrive/Eddie_data/01:2021/'

# Get a list of all nc files in the directory
nc_files = glob.glob(os.path.join(eddy_data_dir, '*.nc'))

# Iterate through each nc file
for file_path in nc_files:
    try:
        ds = xr.open_dataset(file_path)

        # Assuming 'latitude' and 'longitude' are the coordinate variables
        latitude = ds['latitude']
        longitude = ds['longitude']

        # Find indices within the Florida Keys bounding box
        lat_indices = np.where((latitude >= lat_min) & (latitude <= lat_max))[0]
        lon_indices = np.where((longitude >= lon_min) & (longitude <= lon_max))[0]

        # Iterate through the time dimension (assuming 'time' is the time variable)
        if 'time' in ds.variables:
            for t in range(len(ds['time'])):
                # Select the data for the current time step
                ds_time = ds.isel(time=t)

                # Check for 'label_anti' and 'label_cyclo' within the Florida Keys region
                if 'label_anti' in ds_time.variables:
                    label_anti_data = ds_time['label_anti'].values

                    # Extract eddy labels within the Florida Keys region
                    anti_eddy_labels = label_anti_data[np.ix_(lat_indices, lon_indices)]

                    # Get unique non-zero labels (0 typically represents no eddy)
                    unique_anti_labels = np.unique(anti_eddy_labels[anti_eddy_labels != 0])

                    if len(unique_anti_labels) > 0:
                        print(f"File: {os.path.basename(file_path)}, Time step: {t}")
                        print("Anti-cyclonic eddy labels in Florida Keys:", unique_anti_labels)

                if 'label_cyclo' in ds_time.variables:
                    label_cyclo_data = ds_time['label_cyclo'].values

                    # Extract eddy labels within the Florida Keys region
                    cyclo_eddy_labels = label_cyclo_data[np.ix_(lat_indices, lon_indices)]

                    # Get unique non-zero labels (0 typically represents no eddy)
                    unique_cyclo_labels = np.unique(cyclo_eddy_labels[cyclo_eddy_labels != 0])

                    if len(unique_cyclo_labels) > 0:
                        print(f"File: {os.path.basename(file_path)}, Time step: {t}")
                        print("Cyclonic eddy labels in Florida Keys:", unique_cyclo_labels)


        ds.close()
    except Exception as e:
        print("Error processing file {file_path}: {e}")
```
